```python
import pandas as pd 
import numpy as np 
```

```python
fuel_data = pd.read_csv(r'C:\Users\PRohr\Downloads\T_F41SCHEDULE_P52.csv', 
                        usecols=['TOTAL_AIR_HOURS', 'AIR_FUELS_ISSUED','AIRCRAFT_TYPE'])
ac_data   = pd.read_csv(r"C:\Users\PRohr\Downloads\L_AIRCRAFT_TYPE.csv")
```

```python
fuel_data = fuel_data.dropna()
fuel_data = fuel_data[fuel_data.AIR_FUELS_ISSUED != 0]
fuel_data_grouped = fuel_data.groupby(['AIRCRAFT_TYPE'], axis=0).sum()
```

```python
ac_fuel = pd.merge(fuel_data_grouped, ac_data, left_on='AIRCRAFT_TYPE', right_on='Code')
ac_fuel
```

```python
ac_fuel = ac_fuel.drop(['Code'], axis=1)
```

```python
ac_fuel['Fuel per hour'] = ac_fuel['AIR_FUELS_ISSUED']/ac_fuel['TOTAL_AIR_HOURS']
ac_fuel
```
